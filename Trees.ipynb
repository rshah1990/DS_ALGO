### Binary Tree

- A binary tree is a hierarchical data structure in which each node has at most two children generally referred as left child and right child
- **Full binary tree**:  It is a tree in which every node in the tree has either 0 or 2 children
- **Perfect binary Tree**: It is a binary tree in which all interior nodes have two children and all leaves have the same depth or same level.
- **Complete binary tree**: It is a binary tree in which every level, except possibly the last, is completely filled, and all nodes are as far left as possible.
- **Balanced binary tree**: A binary tree is height balanced if it satisfies the following constraints:
    - The left and right subtrees' heights differ by at most one, AND
    - The left subtree is balanced, AND
    - The right subtree is balanced

**unbalanced binary Tree**


- In-order Traversal
    - In this traversal method, the left subtree is visited first, then the root and later the right sub-tree.
    - If a binary tree is traversed in-order, the output will produce sorted key values in an ascending order.
    - D  B  E  A  F  C  G
- Pre-order Traversal
    - In this traversal method, the root node is visited first, then the left subtree and finally the right subtree.
    - A  B  D  E  C  F  G
- Post-order Traversal
    - In this traversal method, the root node is visited last, hence the name. First we traverse the left subtree, then the right subtree and finally the root node.
    - D  E  B  F  G  C  A

### Binary tree with no rule

- In this section we will just search a value in binary tree & print a tree using helper function
- note **that we dont have any rule attached to this tree** 
- it is just to understad how we can use recursion & tree together to search & print tree 

```python
class Node(object):
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None
```

```python
class Tree(object):
    def __init__(self,root):
        self.root = Node(root)
        
    def search(self,find_val):
        return self.search_helper(self.root,find_val)
        
    def search_helper(self,current,find_val):
        if current:
            if current.value == find_val:
                return True
            else:
                return self.search_helper(current.left,find_val) or self.search_helper(current.right,find_val)
        return False
    
    def pre_order_print(self):
        return self.print_helper(self.root)
    
    def print_helper(self,current):
        if current:
            print(current.value)
            self.print_helper(current.left)
            self.print_helper(current.right)
```

```python
tree = Tree(1)
```

```python
tree.root.left = Node(2)
tree.root.right = Node(3)
tree.root.left.left = Node(4)
tree.root.left.right = Node(5)
```

```python
tree.search(2)
```

```python
tree.search(52)
```

```python
tree.pre_order_print()
```

### Binary search Tree

- binary search tree is arranged in perticular order. BST are sorted
- every value left of any node is less than node value & right side value is larger than node value
- search & inserting in binary search tree is O(logn)
- Deleteing a  value is tricky 
- unbalanced binary tree is the worst case scenario. search , insert & delete takes linear time

```python
class BST(object):
    def __init__(self,root):
        self.root = Node(root)
        
    def insert(self,insert_val):
        return self.insert_helper(self.root,insert_val)
    
    def insert_helper(self,current,insert_val):
        if current:
            if current.value < insert_val:
                if current.left:
                    self.insert_helper(current.left,insert_val)
                else:
                    current.left = Node(insert_val)
            else:
                if current.right:
                    self.insert_helper(current.right,insert_val)
                else:
                    current.right = Node(insert_val)
    
```

```python
# Set up tree
tree = BST(4)
```

```python
# Insert elements
tree.insert(2)
tree.insert(1)
tree.insert(3)
tree.insert(5)
```
