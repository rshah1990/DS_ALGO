```python
def fib(n):
    sum = 0
    a = 0
    b = 1
    for i in range(2,n+1,1):
        c = a + b
        a = b
        b = c
        if i%2 == 0:
            sum = sum + c
    return b,sum
```

```python
fib(9)
```

```python
# time complexity = o(n)
# space cmplexity = o(1)
```
